```python
import matplotlib.pyplot as plt
```

```python
week_day = [1,2,3,4,5,6,7]
temp = [50,51,52,48,47,49,46]

plt.plot(week_day,temp)
```

```python
plt.title('weekday vs temperature')
plt.xlabel('weekday')
plt.ylabel('temperature')
plt.plot(week_day,temp)
```

```python
plt.plot(week_day,temp,color='green',linestyle = '--',marker = 'o')
```

```python
plt.plot(week_day,temp,color='cyan',linestyle = '-.',marker = 'D')
```

```python
days = [1,2,3,4,5,6,7]
max_t = [89,88,86,85,84,85,88]
min_t = [54,54,34,5,2,2,45]
avg_t = [12,34,56,7,86,83,56]

plt.plot(days,max_t,label = 'max')
plt.plot(days,min_t,label = 'min')
plt.plot(days,avg_t,label = 'avg')
plt.title('weekday vs temperature')
plt.xlabel('weekday')
plt.ylabel('temperature')
plt.legend()
plt.grid()
```

```python
blood_sugar = [113,95,76,90,98,120,134,157,170,400,234]
plt.hist(blood_sugar)
```

```python
blood_sugar_women = [67,88,34,57,86,87,98]
blood_sugar_men = [67,57,86,43,67,89,24]

plt.hist([blood_sugar_women,blood_sugar_men],bins = [70,100,150,170,200],color =['green','orange'],label = ['women','men'])
plt.legend()
```
